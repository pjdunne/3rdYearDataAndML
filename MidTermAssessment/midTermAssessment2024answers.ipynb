1. In each case describe in one or two sentences what they do. [Marks 1]


*your answer here*

1. The hyperparameters of the k-NN algorithm are
    * number of neighbours $k$
    * the function used to define distance 
    * the weights = `uniform` or `distance`

2. The hyperparameters do the following. 
    * The k parameters is the number of nearest used to assign a class to each point. The class of any point is the most frequent class of its $k$-nearest neighbours,
    * There are many ways to define distance of which the familiar Euclidean metric is the most common.
    * You can also choose how the k-nearest points are weighted when deciding on the class of a point. Common choices are:- 
        * `uniform`, where all k points are considered equal. 
        * `distance`, which weights the contribution of a given training point inversely to the distance from the prediction point i.e. closer training points contribute more.

### 2.1.ii

Name three performance metrics you could use to test the effectiveness of your k-NN algorithm. [Marks 3]



*your answer here*

This is for a binary classification which we have here with the classes being 0 and 1. There are a large number of possible metrics, for example see the [Scikit Learn User Manual](https://scikit-learn.org/stable/modules/model_evaluation.html).  Some of the metrics covered in the course are,

* Accuracy: the proportion of samples that are classified correctly.

* Precision: the proportion of points with predicted to have class 1 that have true class 1 i.e. fraction of predictions of class 1 that are correct.

* Recall or True Positive Rate: fraction of points with true class 1 that are correctly identified as class 1.

* Fall-out or False Positive Rate: the proportion of points with class 0 that are incorrectly predicted as class 1.

* Receiver Operating Characteristic (ROC) curve and area-under-curve (ROC-AUC) metric. Plotting Reall (true positives) against Fall-out (false positives) as we change the decision boundary between 0 and 1 we get a look at the trade-off between these two metrics. The area under this curve gives us a single metric.

````python
# import libraries needed

# see week 2
import pandas as pd

# see week 5
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# see week 6
from sklearn.metrics import accuracy_score, precision_score, recall_score

# see week 7
from sklearn.metrics import roc_curve 
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay

````

````python
# Split data into training and test sets 

# Take 20% of the data out for testing
seed = 0 
test_size_value = 0.2
X_train, X_test, y_train, y_test = train_test_split(coord, 
                                                    label,
                                                    test_size = test_size_value,
                                                    random_state = seed) 
print(y_train.shape,y_test.shape)
````

````python
# Define your hyperparameters

# These are hyperparameters of the k-NN algorithm
neighbours = 4
weight_type = 'uniform'
# weights = 'distance'

````

````python
# Define your model and fit your model to appropriate data

# Now train the model on our training data:
knn_model = KNeighborsClassifier(neighbours, weights = weight_type)
knn_model.fit(X_train,y_train)
````

````python
# Predict on test data 

y_pred = knn_model.predict(X_test)
````

````python
# Use performance metrics to investigate the effectiveness of predictions made
# This is from Week 7

# Accuracy: 
# the proportion of samples that are classified correctly
accuracy=accuracy_score(y_test, y_pred)

# Precision: 
# the proportion of points with predicted to have class +1 that have true class +1 
# i.e. fraction of predictions of class +1 that are correct
precision = precision_score(y_test, y_pred)

# Recall or True Positive Rate: 
# fraction of points with true class +1 that are correctly identified as class +1
recall = recall_score(y_test, y_pred)

#print("Accuracy = {:5.3f}, precision  = {:5.3f}, recall = {:5.3f}, ".format(accuracy,precision,recall ))
print("Accuracy = {:4.1f}%, precision  = {:4.1f}%, recall = {:4.1f}%, ".format(accuracy*100, precision*100, recall*100 ))


````

````python
# This is ROC for k-NN, for example see
# https://stackoverflow.com/questions/52910061/implementing-roc-curves-for-k-nn-machine-learning-algorithm-using-python-and-sci
# For k-NN need the following function
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier.predict_proba
# The threshold used in this ROC-AUC is based on the probability to assign a given test result as +1 class
# y_score_all[n,c] is the fraction of the k-neighbours that in class c.
y_score_all = knn_model.predict_proba(X_test)
y_score = y_score_all[:,1]
#y_score_all.shape

fpr, tpr, thresholds = roc_curve(y_test, y_score) # Find ROC curve values

plt.plot(fpr, tpr) # Plot the ROC curve
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')

k_nn_roc_auc_score = roc_auc_score(y_test, y_score)
print("K-nn with {:d} neighbours and weight type {:s} has ROC-AUC score of {:5.3f}".format(neighbours,weight_type,k_nn_roc_auc_score))
````

````python
# Here it is all together covering several hyperparameter values

# These are hyperparameters of the k-NN algorithm
neighbour_list = range(2,16)
weights_list = ['uniform','distance']

# df = pd.DataFrame(columns = ["weights","neighbours","accuracy"] )
# Warning that df.append is deprecated so try a dictionary of lists
# Fine for a small number of entries.
# Note:- a classic use of a dataframe not an array 
#        as this conatins a mixture of data types.
result_dict = {  "weights": [], 
                 "neighbours": [],
                 "accuracy":[],
                 "precision":[],
                 "recall":[],
                 "roc_auc":[]}
for weight_type in weights_list:
    for neighbours in neighbour_list:
        # Now train the model on our training data:
        knn_model = KNeighborsClassifier(neighbours, weights = weight_type)
        knn_model.fit(X_train,y_train)

        y_pred = knn_model.predict(X_test)
        
        # Now test predictions
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        y_score_all = knn_model.predict_proba(X_test)
        y_score = y_score_all[:,1]
        roc_auc_score_value = roc_auc_score(y_test, y_score)
        """
        print("k-nn: weights="+weight_type
              + ", k={:2d}".format(neighbours) 
              + "   accuracy={:4.1f}%".format(accuracy*100))
        """
        result_dict["weights"].append(weight_type)
        result_dict["neighbours"].append(neighbours)
        result_dict["accuracy"].append(accuracy)
        result_dict["precision"].append(precision)
        result_dict["recall"].append(recall)
        result_dict["roc_auc"].append(roc_auc_score_value)
        #df.append(results, ignore_index=True)
df = pd.DataFrame(result_dict)
display(df[0:9])
````

````python
one_point_accuracy = 1.0/(test_size_value*number_data_points)
print("*** Each data point represents a change of {:f} in the accuracy".format(one_point_accuracy))
````

````python
# code to plot the effectiveness of the results

performance_metric_list = ["accuracy","precision","recall","roc_auc"]

#
number_metrics=len(performance_metric_list)
plot_y = 6*number_metrics 
colour_list=['#D55E00','#56B4E9']
marker_list=['s', '^']
fig, axs = plt.subplots(number_metrics,1,figsize = (10,plot_y),dpi = 150)
for mmm in range(number_metrics):
    performance_metric = performance_metric_list[mmm]
    for iii in range(len(weights_list)):
        weight_type = weights_list[iii]
        axs[mmm].plot(df[df.weights==weight_type]["neighbours"], 
                df[df.weights==weight_type][performance_metric],
                   color=colour_list[iii], 
                   marker=marker_list[iii], 
                    label=weight_type)

    axs[mmm].set_xlabel('$k$',fontsize = 16)
    axs[mmm].set_ylabel(performance_metric,fontsize = 16)
    axs[mmm].tick_params(which='both',labelsize= 12,direction='in',top=True,right=True)
    axs[mmm].grid(color='xkcd:dark blue',alpha = 0.2)
    axs[mmm].xaxis.set_major_locator(MultipleLocator(5))
    axs[mmm].xaxis.set_minor_locator(MultipleLocator(1))
    #ax.yaxis.set_major_locator(MultipleLocator(0.5))
    #ax.yaxis.set_minor_locator(MultipleLocator(0.1))
    axs[mmm].legend(loc='lower right',fontsize = 12,framealpha = 1)
````

### Q3.1

1. What are the two most important hyperparameters of the SVC (support Vector Classification) method?

*your answer here*

Information comes from the [SVC manual page](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC) and/or from Week 6 notebook on SVR (for example Exercise 2 lists them).

* `kernel`: The most important hyperparameter is the kernel to be used as allows non-linear boundaries to be defined. The kernel function $k(\mathbf{\mathbf{X},\mathbf{Y}})$ is where we replace the linear scalar product $\mathbf{X}_i^\intercal\cdot \mathbf{X}_j$  by $k(\mathbf{X}_i,\mathbf{X}_j) = \phi(\mathbf{X}_i)^\intercal\cdot\phi(\mathbf{X}_j)$.
* $C$: The next most important hyperparameter for SVC is a regularisation value $C$. 
    * Much like the soft-margin condition in classification, predictions further than $\epsilon$ from the training $y$ value are penalised according to how much further from the training value they are. Again the scale of this penalisation is controlled by the parameter $C$.

````python
# Here it is all together covering several hyperparameter values. 
# Uses https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
# Code here based on earlier LinearSVC solution

# Now try different hyperparameter values
# Define hyperparameters for SVR

# kernel types:- ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']

kernel_list = ['rbf','poly','linear']
parameter_list = []

if 'linear' in kernel_list:
    kernel_type = 'linear'
    # C values make no difference
    C_list = [0.25*(2**n) for n in range(0,4)]
    for C in C_list:
        parameter_list.append({"kernel":kernel_type, "C":C})

"""
* Polynomial: $k(\mathbf{X}_i,\mathbf{X}_j) = (\gamma\mathbf{X}_i^\intercal \cdot \mathbf{X}_j + r)^d$, 
where $d$ is the order of the polynomial, 
$\gamma$ is a scale parameter and $r$ is a bias parameter. 
$d$ = 1 corresponds to the linear case.
"""        
if 'poly' in kernel_list:
    kernel_type = 'poly'
    # C values make no difference
    C_list = [0.125*(2**n) for n in range(0,4)]
    # gamma='scale' by default. No obvious scaling worth trying here.
    gamma_list = ['scale']# #['scale',0.1,1.0,10.0]
    for C in C_list:
        for gamma in gamma_list:
            parameter_list.append({ "kernel":kernel_type,
                                    "C":C,
                                    "degree":2, 
                                    "gamma":gamma})

"""
* Gaussian radial basis function: 
$k(\mathbf{X}_i,\mathbf{X}_j) = \exp\left(-\gamma||\mathbf{X}_i - \mathbf{X}_j||^2\right)$, 
where $\gamma$ is a scale parameter and $\gamma$ > 0
"""
if 'rbf' in kernel_list:
    kernel_type = 'rbf'
    C_list = [0.125*(2**n) for n in range(0,6)]
    # gamma='scale' by default. No obvious scaling worth trying here.
    gamma_list = [0.5*(2**n) for n in range(0,8)]
    #gamma_list.append('scale')

    for C in C_list:
        for gamma in gamma_list:
            parameter_list.append( { "kernel":kernel_type,
                                     "C":C,
                                     "degree":2, 
                                     "gamma":gamma} )

result_dict = {}
for parameter_dict in parameter_list:
    for key, value in parameter_dict.items():
        result_dict[key]=[] # very lazy code
for output_value in ["accuracy","precision", "recall", "roc_auc"]:
    result_dict[output_value]=[]
           
            

for parameter_dict in parameter_list:
    sss= "*** Trying SVC with: "
    for key, value in parameter_dict.items():
        sss=sss+"   "+key+"="+str(value)
        result_dict[key].append(value)
    print(sss)
    
    svc = SVC(**parameter_dict )
    try:
        svc.fit(X_train, y_train)
    except ConvergenceWarning: 
        print('    FAILED Convergence warning')
        continue
    y_pred = svc.predict(X_test)
    y_score = svc.decision_function(X_test)
    # Now test predictions
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    y_score_all = knn_model.predict_proba(X_test)
    y_score = y_score_all[:,1]
    roc_auc_score_value = roc_auc_score(y_test, y_score)

    #result_dict["kernel"].append(kernel)
    #result_dict["C"].append(C_value)
    result_dict["accuracy"].append(accuracy)
    result_dict["precision"].append(precision)
    result_dict["recall"].append(recall)
    result_dict["roc_auc"].append(roc_auc_score_value)
    print("     Accuracy {:5.3f}, ROC-AUC {:5.3f}".format(accuracy,roc_auc_score_value))
    #df.append(results, ignore_index=True)

df = pd.DataFrame(result_dict)
display(df[0:9])

````
